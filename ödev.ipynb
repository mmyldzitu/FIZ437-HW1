```python
import numpy as np

import matplotlib.pyplot as plot
import math
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

time= np.arange(0,1,0.01).reshape(-1,1)


mean=0
variance=0.05

amplitude   = np.sin(2*math.pi*time)+np.random.normal(mean,variance)

def x_eksenitrain(a):
    z=a*80/100
    time_2=np.sort(np.random.rand(int(z),1),axis=0)
    return time_2
def y_eksenitrain(b):
    z=b*80/100
    amplitude_2   = np.sin(2*math.pi*b)+np.random.normal(mean,variance)
    return amplitude_2

def x_eksenitest(a):
    z=a*20/100
    time_3=np.sort(np.random.rand(int(z),1),axis=0)
    return time_3

def y_eksenitest(b):
    
    amplitude_3  = np.sin(2*math.pi*b)+np.random.normal(mean,variance)
    return amplitude_3

    
 
mylist=[5,10,20,50,100,1000]
for i in mylist:
    print(i)
    polynomial_regression = PolynomialFeatures(degree = 10) 
    x_pol = polynomial_regression.fit_transform(x_eksenitrain(i))
    x_pol_2 = polynomial_regression.fit_transform(x_eksenitest(i))
    linear_regression=LinearRegression()
    linear_regression.fit(x_pol,y_eksenitrain(x_eksenitrain(i)))
    linear_regression.fit(x_pol_2,y_eksenitest(x_eksenitest(i)))
    
    plot.plot(x_eksenitrain(i),y_eksenitrain(x_eksenitrain(i)),color= "green",label = "train")
    plot.plot(x_eksenitest(i),y_eksenitest(x_eksenitest(i)),color= "red",label = "test")
    plot.plot(time,amplitude,label="ground truth")
    plot.legend()
    plot.show()
    
    
   









```
